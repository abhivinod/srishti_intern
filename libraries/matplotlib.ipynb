```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='green',linestyle='--',marker='o')
plt.title('Plot btwn x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,8,27,64,81]
plt.plot(x,y1,label='Square',marker='o')
plt.plot(x,y2,label='Cubic',marker='o')
plt.legend()
plt.show()
```

```python
categories=['A','B','C']
values=[5,7,3]
plt.bar(categories,values)
plt.show()
```

```python
data=[1,1,2,3,3,3,4,4,5,1,2]
plt.hist(data,bins=5)
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.scatter(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[1,4,9])
plt.subplot(1,2,2)
plt.plot([1,2,3],[1,8,27])
plt.show()
```

```python
plt.plot(x,y)
plt.savefig('plot.png')
```
