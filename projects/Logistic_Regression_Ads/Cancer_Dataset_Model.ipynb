```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
cancer=load_breast_cancer()
x_cancer=cancer.data
y_cancer=cancer.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_cancer,y_cancer)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_test
```

```python
classifier=LogisticRegression(max_iter=1000).fit(x_train,y_train)
```

```python
import pandas as pd
y_pred=classifier.predict(x_test)
print(pd.DataFrame({"Predicted":y_pred,"Actual":y_test}))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm,"\n",accuracy_score(y_test,y_pred),"\n",classification_report(y_test,y_pred))
```

```python
import joblib,pickle
joblib.dump(classifier,"cancer_model.joblib")
with open('cancer_model.pkl','wb') as f:
    pickle.dump(classifier,f)
```
