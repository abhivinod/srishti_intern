```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
X=dataset.iloc[:,:-1].values #.values used to extract only the value/data as a numpy array
Y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(x_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
print(x_test)
```

```python
print(classifier.predict([[-0.0855487,0.22453427]]))
```

```python
print(classifier.predict(sc.transform([[50,40000]])))
```

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_pred),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
DecisionBoundaryDisplay.from_estimator(
    classifier,
    x_train,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1],)
plt.scatter(x_train[:,0],x_train[:,1],
            c=y_train,
            s=20,
            edgecolors="k")
plt.title('Logistic Regression for (Training Set)')
plt.show()
```

```python
DecisionBoundaryDisplay.from_estimator(
    classifier,
    x_test,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1],
)
plt.scatter(x_test[:,0],x_test[:,1],
    c=y_test,s=20,edgecolors="k")
plt.title("Logistic Regression(Testing set)")
plt.show()
```

```python
import joblib
joblib.dump(classifier,"svm_model.joblib")#saving
svm_model=joblib.load("svm_model.joblib")
svm_model.predict(sc.transform([[50,40000]]))
```
