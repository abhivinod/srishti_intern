```python
#Alternative Method
from sklearn.datasets import load_iris
iris_data=load_iris()
iris.head()
```

```python
iris_data['data']#loads a 150x4 size data frame ,value of sepal,petal lengths and widths of 150 flowers
iris_data['target']#Loads a 1-d data frame having values 0,1,2 for 3 iris species
iris_data['target_names']#Loads the names of target flowers-setosa,versicolor,virginica
iris_data['DESCR']#Describes the dataset
iris_data['feature_names']#Names of features,sepal,petal width,lengths
iris_data['filename']#filename
iris_data['data_module']
```

```python
iris_df=pd.DataFrame(data=iris_data['data'],columns=iris_data['feature_names'])
iris_df.head()
```

```python
iris_df['Flower']=iris_data['target_names'][iris_data['target']]
iris_df.head()
```

```python
iris_df.rename(columns={"sepal length (cm)":"sepal_length"},inplace=True)
iris_df.rename(columns={"sepal width (cm)":"sepal_width"},inplace=True)
iris_df.rename(columns={"petal length (cm)":"petal_length"},inplace=True)
iris_df.rename(columns={"petal width (cm)":"petal_width"},inplace=True)
iris_df.head()
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,10))
#For sepal length
sns.histplot(iris_df['sepal_length'],kde=True,ax=axes[0,0],color='green')
axes[0,0].set_title('Sepal Length Distribution')
sns.histplot(iris_df['sepal_width'],kde=True,ax=axes[0,1],color='red')
axes[0,1].set_title('Sepal Width Distribution')
sns.histplot(iris_df['petal_length'],kde=True,ax=axes[1,0],color='blue')
axes[1,0].set_title('Petal Length Distribution')
sns.histplot(iris_df['petal_width'],kde=True,ax=axes[1,1],color='yellow')
axes[1,1].set_title("Petal Width Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
species=iris_df['Flower'].unique()
colors=['red','green','blue']
for i in range(len(colors)):
    z=iris_df[iris_df['Flower']==species[i]]
    plt.scatter(x=z['sepal_length'],y=z['sepal_width'],c=colors[i],label=species[i])
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Sepal Width(cm)')
plt.title('Sepal Length vs Sepal Width by Species')
plt.legend()
plt.show()
```
