```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
dataset=load_iris()
print(dataset.keys())
```

```python
dataset.target
```

```python
data_arr=pd.DataFrame(np.concat((dataset["data"],dataset["target"].reshape(-1,1)),1),columns=dataset['feature_names'].append(dataset.target))
```

```python
data_arr
```

```python
X=data_arr.iloc[:,0:4].values
Y=data_arr.iloc[:,4].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0)
```

```python
print(x_train.size,x_test.size)
```

```python
model=tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
print(pd.DataFrame(np.hstack((x_test,y_test.reshape(-1,1)))))
```

```python
y_predicted=model.predict(x_test)
```

```python
print(np.concat((y_predicted.reshape(-1,1),y_test.reshape(-1,1)),1))
```

```python
#Evaluation
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(f"{confusion_matrix(y_predicted,y_test)}\n{accuracy_score(y_predicted,y_test)}\n{classification_report(y_predicted,y_test)}")
      
```

```python
#visualize tree structure usingf graphviz
import graphviz
dot_data=tree.export_graphviz(
    model,
    out_file=None,
    feature_names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],
    class_names=[str(s) for s in dataset.target_names],
    filled=True,
    rounded=True,
    special_characters=True)
graph=graphviz.Source(dot_data)
graph
```
