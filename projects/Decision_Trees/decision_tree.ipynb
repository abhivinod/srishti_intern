```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0) #Split dataset into training and testing data
```

```python
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier()
model1.fit(x_train,y_train)
```

```python
model2=DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
model2.fit(x_train,y_train)
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    np.hstack((x_train[:, :2], y_train.reshape(-1, 1))),
    columns=["Feature1", "Feature2", "Target"]
)

print(df.head())
```

```python
print(model1.predict([[0.797535,-1.229939]]))
print(model2.predict([[0.797535,-1.229939]]))
```

```python
#Predicting with default model and comparing with actual target
predict1=model1.predict(x_test)
print(np.hstack((predict1.reshape(-1,1),y_test.reshape(-1,1))))
```

```python
#Predicting with tuned model and comparing with actual target
predict2=model2.predict(x_test)
print(np.concat((predict2.reshape(-1,1),y_test.reshape(-1,1)),1))
```

```python
#Evaluation of defualt model
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(predict1,y_test))
print(accuracy_score(predict1,y_test))
print(classification_report(predict1,y_test))
```

```python
#Evaluation of tuned model
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(predict2,y_test))
print(accuracy_score(predict2,y_test))
print(classification_report(predict2,y_test))
```

```python
#visualising training set results for default model
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
DecisionBoundaryDisplay.from_estimator(
    model1,
    x_train,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1])

plt.scatter(x_train[:,0],x_train[:,1],
            c=y_train,
            s=20,edgecolors="k")
plt.title('Decision Tree-Classifier(Training set),Default Model')
plt.show()
```

```python
#visualising training set results for tuned model
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
DecisionBoundaryDisplay.from_estimator(
    model2,
    x_train,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1])

plt.scatter(x_train[:,0],x_train[:,1],
            c=y_train,
            s=20,edgecolors="k")
plt.title('Decision Tree-Classifier(Training set),Tuned Model')
plt.show()
```

```python
#visualising testing set results for default model
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
DecisionBoundaryDisplay.from_estimator(
    model1,
    x_test,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1])

plt.scatter(x_test[:,0],x_test[:,1],
            c=y_test,
            s=20,edgecolors="k")
plt.title('Decision Tree-Classifier(Testing set),Default Model')
plt.show()
```

```python
#visualising testing set results for tuned model
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
DecisionBoundaryDisplay.from_estimator(
    model2,
    x_test,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1])

plt.scatter(x_test[:,0],x_test[:,1],
            c=y_test,
            s=20,edgecolors="k")
plt.title('Decision Tree-Classifier(Testing set),Tuned Model')
plt.show()
```

```python
#Function for displaying the tree structure
from sklearn import tree
def plot_decision_tree(model,feature_names,class_names):
    plt.figure(figsize=(15,10))
    tree.plot_tree(model,  #trained model
                    filled=True,#Nodes are coloured based on the class
                    feature_names=feature_names,
                    class_names=class_names,
                    rounded=True) 
    plt.show()
```

```python
#Tree structure for default model
plot_decision_tree(model1,['Age','EstimatedSalary'],['Not Purchased','Purchased'])
```

```python
#Tree structure for tuned model
plot_decision_tree(model2,['Age','EstimatedSalary'],['Not Purchased','Purchased'])
```

```python
imp=model2.feature_importances_
print("Feature importances for tuned model:")
for name,importance in zip(['Age','EstimatedSalary'],imp):
    print(f"{name}:{importance:.5f}")
```
