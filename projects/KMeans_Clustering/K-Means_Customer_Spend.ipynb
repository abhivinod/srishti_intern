```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
data=dataset.iloc[:,[3,4,]].values
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Number of Cluster(Elbow Method)")
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
kmeans.fit(data)
y_kmeans=kmeans.predict(data)
```

```python
y_kmeans
```

```python
#Visualize using scatterplot
plt.scatter(data[y_kmeans==0,0],data[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(data[y_kmeans==1,0],data[y_kmeans==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(data[y_kmeans==2,0],data[y_kmeans==2,1],s=100,c='green',label='Cluster3')
plt.scatter(data[y_kmeans==3,0],data[y_kmeans==3,1],s=100,c='magenta',label='Cluster4')
plt.scatter(data[y_kmeans==4,0],data[y_kmeans==4,1],s=100,c='cyan',label='Cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
